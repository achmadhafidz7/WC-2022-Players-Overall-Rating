## Import Libraries and return status code of the website

```python
import pandas as pd
import requests

pd.options.mode.chained_assignment = None  # default='warn'
wikiurl = "https://en.wikipedia.org/wiki/2022_FIFA_World_Cup_squads"
```

```python
# make request to website and return the status code
# status code 200 = it's allowed to scrape

response = requests.get(wikiurl)
print(response.status_code)
```

## Create a variable that contains all the table from website using pd.read_html()

```python
tables = pd.read_html(wikiurl)
print(tables)
```

## Concat all the tables needed into one dataframe

```python
# Check how many tables are there

print(len(tables))
```

```python
tables_concat = pd.concat(tables[:32])
df = tables_concat[['No.', 'Pos.', 'Player','Club']]
df
```

## Show all the players to check their index for inserting nationality

```python
# reset the index
df.reset_index(drop=True, inplace=True)
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
display(df)
```

## Create a function to insert nationality of each players

```python
def nat(a,b,c):
    df.loc[a:b,'Nationality'] = c
```
